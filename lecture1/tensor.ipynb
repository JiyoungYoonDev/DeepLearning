```python
import torch
```

```python
a = torch.tensor([1, 2, 3])
print(a.ndim)
print(a.dtype)
```

```python
a = torch.tensor([1, 2, 3], dtype=torch.float32)
print(a)
```

```python
x=torch.tensor([[1,2,3], [4,5,6]])
print(x.ndim)
print(x.shape)
print(x.dtype)
```

```python
y=torch.tensor([1.0,2.0,3.0])
print(y.dtype)

z=torch.tensor(5)
print(z.ndim)
print(z.shape)
```

# Zeros function
- we can also use .rand, .ones

```python
b = torch.zeros(3, dtype=torch.float32)
print(b)

b = torch.zeros([10, 10], dtype=torch.float32)
print(b)

b = torch.zeros([3, 3, 3], dtype=torch.float32)
print(b)
```

```python
a = torch.arange(10)
print(a)
print(f'{a=}')
```
