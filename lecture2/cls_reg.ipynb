nn.Linear는 파이토치에서 선형 회귀 공식인 $y = Wx + b$ 연산을 내부적으로 알아서 수행하고 가중치($W$)와 편향($b$)을 자동으로 생성·관리해 주는 가장 기본적인 레이어입니다.

```python
import torch
import torch.nn as nn
```

```python
model = nn.Linear(10, 1)
```

```python
print(model)
print(model.weight)
print(model.bias)
```
